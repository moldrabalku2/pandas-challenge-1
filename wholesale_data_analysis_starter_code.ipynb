## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

# Created the 'line_subtotal' column by multiplying 'unit_price' by 'qty'
df_transform = df
df_transform['line_subtotal'] = df['unit_price'] * df['qty']

# Output 'unit_price', 'qty', and 'line_subtotal'
result_df = df_transform[['unit_price', 'qty', 'line_subtotal']]
result_df.head(2)
```

```python
# Create a column for shipping price.
#
# Assume a shipping price of $7 per pound for orders 
# over 50 pounds and $10 per pound for items 50 pounds or under.
#
# note to self we are using unit_weight	qty then if the weight 
# is over 50 lbs its $7 per lbs if under its $10 per lbs 
# going to do this with a for loop
# Initialize an empty list to store the shipping prices
df_transform = df
shipping_prices = []

#Loop through each row of the DataFrame
for index, row in df.iterrows():
    total_weight = row['unit_weight'] * row['qty']
    
    #Calculate the shipping price based on the weight condition
    if total_weight > 50:
        shipping_price = total_weight * 7  # $7 per pound for orders over 50 lbs
    else:
        shipping_price = total_weight * 10  # $10 per pound for orders 50 lbs or under
    
    # Append the shipping price to the list
    shipping_prices.append(shipping_price)

# Assign the calculated shipping prices back to the DataFrame
df_transform['shipping_price'] = shipping_prices

# Output 'unit_price', 'unit_weight', 'qty', 'total_weight', and 'shipping_price'
df_transform['total_weight'] = df_transform['unit_weight'] * df_transform['qty']
result_df_a = df_transform[['unit_price', 'unit_weight', 'qty','total_weight', 'shipping_price']]

# Display the result differentiating this result from the below result using a and b at the end
result_df_a.head(3)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df_transform = df
sales_tax_rate = 0.0925  # 9.25% sales tax

# Caluclulating the subtotal for the line_subtotal and shipping price
df_transform['shipping_line_subtotal'] = df_transform['line_subtotal'] + df_transform['shipping_price']

# Calculate sales tax based on the shipping_line_subtotal
df_transform['sales_tax'] = df_transform['shipping_line_subtotal'] * sales_tax_rate

# Calculate the final line price (subtotal + shipping + sales tax)
df_transform['line_price'] = df_transform['line_subtotal'] + df_transform['shipping_price'] + df_transform['sales_tax']

# Round the 'line_price' to two decimal places
df_transform['line_price'] = df_transform['line_price'].round(2)

# Output the relevant columns: 'line_subtotal', 'shipping_price' and 'line_price'
result_df_b = df_transform[['line_subtotal', 'shipping_price', 'line_price']]

# Display the result differentiating the above result from this by using a and b at the end
result_df_b.head(3)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
# geting the total unit cost using  unit cost times qty
df_transform['unit_cost_total'] = df_transform['unit_cost'] * df_transform['qty']

# Calculate line cost (unit cost total + shipping price)
df_transform['line_cost'] = df_transform['unit_cost_total'] + df_transform['shipping_price']

# Selecting the relevant columns to display
result_df_final = df[['first', 'last', 'job', 'phone', 'email', 'client_id', 'order_id', 'order_week', 
                      'order_year', 'unit_price', 'unit_cost', 'unit_weight', 'qty', 'line_number', 
                      'line_subtotal', 'total_weight', 'shipping_price', 'line_price', 'line_cost']]

# Display the final result first five
result_df_final.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price
# first,last, job, phone, email, client_id,	order_id, order_date, order_week, 
# order_year, unit_cost, unit_weight, qty, line_number,	line_subtotal, total_weight,	
# shipping_price, line_price, line_cost, line_profit
# Calculate the profit for each line (line_price - line_cost)
df_transform['line_profit'] = df_transform['line_price'] - df_transform['line_cost']

# Displaying the following columns
# 'first', 'last', 'job', 'phone', 'email', 'client_id', 'order_id', 'order_week', 
# 'order_year', 'unit_price', 'unit_cost', 'unit_weight', 'qty', 'line_number', 
# 'line_subtotal', 'shipping_price', 'line_price', 'line_cost', 'line_profit'

result_df_profit = df_transform[['first', 'last', 'job', 'phone', 'email', 'client_id', 'order_id', 'order_week', 
                      'order_year', 'unit_price', 'unit_cost', 'unit_weight', 'qty', 'line_number', 
                      'line_subtotal', 'total_weight','shipping_price', 'line_price', 'line_cost', 'line_profit']]

# Display the final result (first five rows)

result_df_profit.head(3)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# Find the top 5 clients by number of entries (using value_counts)
top_5_clients_by_entries = df['client_id'].value_counts().head(5)
top_5_client_ids = top_5_clients_by_entries.index.tolist()

# Initialize an empty list to store summary data for each client
summary_data = []

# Loop through the top 5 clients and calculate the required metrics
for client_id in top_5_client_ids:
    # Filter the DataFrame for each client
    client_data = df[df['client_id'] == client_id]
    
    # Calculate the total quantity purchased by the client
    total_qty = client_data['qty'].sum()
    
    # Calculate the total shipping price for the client
    total_shipping_price = client_data['shipping_price'].sum()
    
    # Calculate the total revenue (line_price) for the client
    total_revenue = client_data['line_price'].sum()
    # Calculate the total line cost for the client
    
    total_line_cost = client_data['line_cost'].sum()
    # Calculate the total profit (line_profit) for the client
    total_profit = client_data['line_profit'].sum()
    
    # Append the results for each client to the summary list
    summary_data.append({
        'client_id': client_id,
        'total_units_purchased': total_qty,
        'total_shipping_price': total_shipping_price,
        'total_revenue': total_revenue,
        'line_cost': total_line_cost,
        'total_profit': total_profit
    })

# Create a summary DataFrame from the summary_data list
summary_df = pd.DataFrame(summary_data)

summary_df.columns = ['client_id', 'qty', 'shipping_price', 
                        'line_price', 'line_cost', 
                        'line_profit']
# Display the summary DataFrame
summary_df

```

```python
# Format the data and rename the columns to names suitable for presentation.
summary_df = pd.DataFrame(summary_data)
summary_df.columns = ['Client ID', 'Total Units Purchased', 'Total Shipping Cost', 'Total Revenue', 'Total Cost', 'Total Profit']
# Define the  columns in millions. 
def currency_format_millions(x):
    return '{:,.6f}'.format(x / 1_000_000)

# Define a function that converts a dollar amount to millions.
summary_df['Total Shipping Cost'] = summary_df['Total Shipping Cost'].apply(currency_format_millions)
summary_df['Total Revenue'] = summary_df['Total Revenue'].apply(currency_format_millions)
summary_df['Total Cost'] = summary_df['Total Cost'].apply(currency_format_millions)
summary_df['Total Profit'] = summary_df['Total Profit'].apply(currency_format_millions)

# Apply the currency_format_millions function to only the money columns. 
summary_df.columns = ['Client ID', 'Total Units Purchased', 'Total Shipping Cost (in Millions)', 
                    'Total Revenue (in Millions)', 'Total Cost (in Millions)', 'Total Profit (in Millions)']

# Rename the columns to reflect the change in the money format. 
summary_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_summary_df = summary_df.sort_values(by='Total Profit (in Millions)', ascending=False)

# Display the sorted DataFrame
sorted_summary_df
```
